```python
def resizeAllImagesInDir(pathToDir):
    for filename in os.listdir(pathToDir):
        abs_filename = os.path.join(pathToDir, filename)
        if os.path.isfile(abs_filename):
            img = cv2.imread(abs_filename)
            if img is not None:
              img = cv2.resize(img, (224, 224))
              cv2.imwrite(abs_filename, img)
```

```python
path1 = '/content/RICK_MORTY/Morty/'
val_cat = [path1 + f for i, f in enumerate(os.listdir(path1)) if i < 500]
train_cat = [path1 + f for i, f in enumerate(os.listdir(path1)) if i >= 500]

path2 = '/content/RICK_MORTY/Rick/'
val_dog = [path2 + f for i, f in enumerate(os.listdir(path2)) if i < 500]
train_dog = [path2 + f for i, f in enumerate(os.listdir(path2)) if i >= 500]

path3 = '/content/RICK_MORTY/trash/'
val_trash = [path3 + f for i, f in enumerate(os.listdir(path3)) if i < 500]
train_trash = [path3 + f for i, f in enumerate(os.listdir(path3)) if i >= 500]
```

```python
os.mkdir('RICK_MORTY/train')
os.mkdir('RICK_MORTY/val')
os.mkdir('RICK_MORTY/train/Rick')
os.mkdir('RICK_MORTY/train/Morty')
os.mkdir('RICK_MORTY/train/trash')
os.mkdir('RICK_MORTY/val/Rick')
os.mkdir('RICK_MORTY/val/Morty')
os.mkdir('RICK_MORTY/val/trash')
```

```python
def divide_directories(path, images):
  for img in images:
    shutil.move(img, path)
```

```python
divide_directories('RICK_MORTY/train/Rick/', train_dog)
divide_directories('RICK_MORTY/train/Morty/', train_cat)
divide_directories('RICK_MORTY/train/trash/', train_trash)
divide_directories('RICK_MORTY/val/Rick/', val_dog)
divide_directories('RICK_MORTY/val/Morty/', val_cat)
divide_directories('RICK_MORTY/val/trash/', val_trash)
```

```python
print(set(train_cat).intersection(set(val_cat)))
print(set(train_dog).intersection(set(val_dog)))
print(set(train_trash).intersection(set(val_trash)))
```
